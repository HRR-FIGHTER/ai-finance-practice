```python
# --- 1. Import libraries ---
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt

# --- 2. Prepare the data ---

# Make sure Date is index
df_last2y.index = pd.to_datetime(df_last2y.index)

# Select only the columns we want: Closing Price (CP) and sentiment_score
features = ["CP", "sentiment_score"]
data = df_last2y[features].copy()

# Scale features to [0,1] for LSTM
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# --- 3. Create sequences (look-back window) ---
def create_sequences(data, seq_length=60):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])   # past seq_length days
        y.append(data[i+seq_length, 0])  # predict CP (column 0)
    return np.array(X), np.array(y)

SEQ_LENGTH = 60  # using 60 past days to predict next day
X, y = create_sequences(scaled_data, SEQ_LENGTH)

# --- 4. Train/Test split ---
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# --- 5. Build LSTM model ---
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH, len(features))),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25, activation='relu'),
    Dense(1)   # predict next day's CP
])

model.compile(optimizer='adam', loss='mse')
model.summary()

# --- 6. Train the model ---
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=20, batch_size=32)

# --- 7. Make predictions ---
predictions = model.predict(X_test)

# Reverse scaling: add zeros for sentiment_score column
predicted_prices = scaler.inverse_transform(
    np.concatenate(
        (predictions, np.zeros((predictions.shape[0], scaled_data.shape[1]-1))),
        axis=1
    )
)[:,0]

# Actual CP values for comparison
actual_prices = df_last2y["CP"].values[-len(predicted_prices):]

# --- 8. Plot results ---
plt.figure(figsize=(12,6))
plt.plot(actual_prices, label="Actual CP")
plt.plot(predicted_prices, label="Predicted CP")
plt.title("LSTM Predictions (CP + Sentiment)")
plt.xlabel("Time")
plt.ylabel("Closing Price (CP)")
plt.legend()
plt.show()
```
